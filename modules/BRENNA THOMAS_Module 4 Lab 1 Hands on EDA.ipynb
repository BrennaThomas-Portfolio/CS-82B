RM: Average number of rooms per dwelling

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load the data
df = pd.read_csv('Cardiovascular_Disease_Dataset.csv', sep=None, engine='python')  
df.columns = df.columns.str.strip() 
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

# Do older people have more cardiovascular disease?

## Distribution of Age by Cardiovascular Disease

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='age', hue='target', bins=30, kde=True, palette='Set1')
plt.title('Age Distribution by Cardiovascular Disease')
plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.legend(['No Disease', 'Disease'])
plt.show()
```

# Is disease more common in a certain gender?

## Frequency Distribution of Gender vs Cardiovascular Disease

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='gender', hue='target', palette='pastel')
plt.title('Gender and Cardiovascular Disease')
plt.xlabel('Gender (0: Male, 1: Female)')  
plt.ylabel('Count')
plt.legend(['No Disease', 'Disease'])
plt.show()
```

# Is high systolic BP associated with disease?



## Boxplot of Resting BP vs Disease

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='target', y='restingBP', hue='target', palette='Set2')
plt.title('Resting Blood Pressure by Disease Status')
plt.xlabel('Cardiovascular Disease (0: No, 1: Yes)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

# How does cholesterol status relate to age and disease risk?



## Violin Plot for Cholesterol vs Age by Disease

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x='target', y='serumcholestrol', hue='target', palette='Set3')
plt.title('Cholesterol Level by Disease Status')
plt.xlabel('Cardiovascular Disease (0: No, 1: Yes)')
plt.ylabel('Serum Cholesterol')
plt.show()
```

# How do age and maximum heart rate relate to cardiovascular disease?



## Max Heart Rate vs Age by Cardiovascular Disease Status

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='age', y='maxheartrate', hue='target', palette='coolwarm')
plt.title('Max Heart Rate vs Age by Disease Status')
plt.xlabel('Age (years)')
plt.ylabel('Max Heart Rate')
plt.legend(['No Disease', 'Disease'])
plt.show()
```

# How do age and cholesterol relate to cardiovascular disease?

##  Cholesterol vs Age, colored by Disease

```python

df['age_years'] = df['age']
```

```python
disease_map = {0: 'No Disease', 1: 'Disease'}
df['disease_status'] = df['target'].map(disease_map)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='age_years',
    y='serumcholestrol',
    hue='disease_status',
    palette='Set1',
    alpha=0.7
)
plt.title('Cholesterol vs Age by Cardiovascular Disease Status')
plt.xlabel('Age (years)')
plt.ylabel('Serum Cholesterol (mg/dL)')
plt.legend(title='Disease Status')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
